# 1. Plotting Two Variables
Take some time going through the code cells below, and deconstruct for yourself what is being done in each case.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 1000)
y = np.cos(x)

# make the plot
plt.plot(x,y)

# Show plot
plt.show()
```

We can display multiple plots on the same figure by placing as many plt.plot() statements as we need.  For example, lets plot \
$y=cos(x)$, 
$y=sin(x)$,
and $y=tan(x)$ on the same figure with the same interval of $0<x<10$.

```python
x = np.linspace(0, 10, 1000)
y = np.cos(x)

# make the plot
plt.plot(x,y)
plt.plot(x,np.sin(x))
```

# 1.1 Labels/Legends
Never forget to label your axes and plots!

```python
x = np.linspace(0, 10, 1000)
y = np.cos(x)

# make the plot
plt.plot(x,y, label = 'cox(x)')
plt.plot(x,np.sin(x), label = 'sin(x)')

# legend
plt.legend()

# axes
plt.ylabel('y')
plt.xlabel('x')
plt.title('title')
```

# 1.2 Color Schemes
You'll notice from the last figure we made that Python automatically picks color schemes for multiple plots on the same figure.  You can manually change the color of each plot by adding a "color" keyword in the plt.plot() function.

```python
x = np.linspace(0, 10, 1000)
y = np.cos(x)

# make the plot
plt.plot(x,y, 'bo', label = 'cox(x)')
plt.plot(x,np.sin(x), 'm', label = 'sin(x)')

# legend
plt.legend()

# axes
plt.ylabel('y')
plt.xlabel('x')
plt.title('title')
```

Picking an effective color scheme is very important for data visualization.  Here is an example of a bad color scheme:

```python
x = np.linspace(0, 10, 1000)
y = np.cos(x)

# make the plot
plt.plot(x,y, 'y', label = 'cox(x)')
plt.plot(x,np.sin(x), color = 'orange', label = 'sin(x)')

# legend
plt.legend()

# axes
plt.ylabel('y')
plt.xlabel('x')
plt.title('title')
```

You can barely tell the difference between the orange and yellow plots,and even the magenta doesn't look quite right.  Picking the right color scheme is especially import considering that people trying to read your plots may be colorblind.  Here is an article on color schemes that I use as a cheat sheet:  https://chartio.com/learn/charts/how-to-choose-colors-data-visualization/

Using an example from the article, a better choice of colors would be:

```python
x = np.linspace(0, 10, 1000)
y = np.cos(x)

# make the plot
plt.plot(x,y, 'r', label = 'cos(x)')
plt.plot(x,np.sin(x), 'b,', label = 'sin(x)')

# legend
plt.legend()

# axes
plt.ylabel('y')
plt.xlabel('x')
plt.title('title')
```

# 2. Plotting Three Variables

At some point, you will likely have to visualize data where you have three variables.  Classically, this can be done by envisioning a 3rd dimension to a plot.  As an example, lets plot a 2-dimensional Gaussian. \\
$z=e^{-x^2-y^2}$

We can start by plotting this as a contour map, where each line represents a constant value, and the color tells you what value specifically is being plotted.

```python
#define range of values
x_gauss = np.linspace(-10, 10, num = 100)
y_gauss = np.linspace(-10, 10, num = 100)

# create grid of x and y
x_gauss, y_gauss = np.meshgrid(x_gauss, y_gauss)

# calculate z
# factor of .1 makes the gaussian wider for visual purposes
z_gauss = np.exp(-.1*x_gauss**2 - .1*y_gauss**2)
cdata = plt.contour(z_gauss)
plt.ylabel('y')
plt.xlabel('x')
plt.title('2D Gaussian')
plt.colorbar(cdata)
```

An alternative to the classical contour plot is an "image" or "heat" map.  Instead of visualizing the 3rd variable using contour lines, we visualize with color.  As an example, lets plot the same 2D Gaussian from before, but as a density map.

```python
# sizing
plt.figure(figsize = (8,8))
#define range of values
x_gauss = np.linspace(-10, 10, num = 100)
y_gauss = np.linspace(-10, 10, num = 100)

# create grid of x and y
x_gauss, y_gauss = np.meshgrid(x_gauss, y_gauss)

# calculate z
# factor of .1 makes the gaussian wider for visual purposes
z_gauss = np.exp(-.1*x_gauss**2 - .1*y_gauss**2)
cdata = plt.imshow(z_gauss)
plt.ylabel('y')
plt.xlabel('x')
plt.title('2D Gaussian')
plt.colorbar(cdata)
```

You'll notice that ever since we made the figure bigger, the text of the plots labels and numbers did not grow with it.  We can manually adjust this as well with 1 line of code, although this sets the default for all future plots in this notebook.

```python
# sizing
plt.figure(figsize = (8,8))
plt.rc('font', size = 20)
#define range of values
x_gauss = np.linspace(-10, 10, num = 100)
y_gauss = np.linspace(-10, 10, num = 100)

# create grid of x and y
x_gauss, y_gauss = np.meshgrid(x_gauss, y_gauss)

# calculate z
# factor of .1 makes the gaussian wider for visual purposes
z_gauss = np.exp(-.1*x_gauss**2 - .1*y_gauss**2)
cdata = plt.imshow(z_gauss)
plt.ylabel('y')
plt.xlabel('x')
plt.title('2D Gaussian')
plt.colorbar(cdata)
```

# 3.2 Scales
Density maps are very useful in  images.  Sometimes there may be features in an image that can only be seen by changing the scale of the image.  This can be done in many different ways in Python, but here is a simple example:

```python
plt.figure(figsize = (8,8))
plt.rc('font', size = 20)

#define range of values
x_gauss = np.linspace(-10, 10, num = 100)
y_gauss = np.linspace(-10, 10, num = 100)

# create grid of x and y
x_gauss, y_gauss = np.meshgrid(x_gauss, y_gauss)

# calculate z
# factor of .1 makes the gaussian wider for visual purposes
z_gauss = np.exp(-.1*x_gauss**2 - .1*y_gauss**2)

vmin = 0
vmax = 0.1

plt.imshow(z_gauss, vmin = vmin, vmax = vmax)
plt.ylabel('y')
plt.xlabel('x')
plt.title('2D Gaussian')

# colorbar
plt.colorbar()
```

This image is actually a good counter example.  With the new scale we implemented, we see the overall structure of the Gaussian, but we've lost the brightness gradient towards the center.

# 4. Subplotting
Sometimes you may wish to have multiple plots side-by-side, stacked on top of one another, or perhaps you want a grid of plots.  we can this "subplotting". 

#4.1 Side-by-side
As an example, lets start with a side-by-side plot.

```python
# define the figure for all of the plots
fig, axs = plt.subplots(1, 2, figsize = (20,8))

# make the first subplot
axs[0].plot(x,y)

#make the second subplot
axs[1].plot(x, np.log(x))
```

# 4.2 Grids

Now lets try making a grid of plots.  This can be a bit confusing if it's your first time seeing it, so I encourage patience with yourself and asking lots of questions if it doesn't make sense.

```python
# define the figure for all of the plots
fig, axs = plt.subplots(2, 2, figsize = (20,8))

# make the first subplot
axs[0,0].plot(x,y)
axs[0,0].set_xlabel('x')

#make the second subplot
axs[0,1].plot(x, np.log(x))

# make the third subplot
axs[1,0].plot(x,y)

#make the fourth subplot
axs[1,1].plot(x, np.tan(x))
```

Great!  But what's all this funny buisness with the overlaid text?  I honestly don't know, but I do know how to fix it!  All you have to do (typically) in this situation is add "plt.tight_layout()" after making the figure.

```python
# define the figure for all of the plots
fig, axs = plt.subplots(2, 2, figsize = (20,8))

# make the first subplot
axs[0,0].plot(x,y)
axs[0,0].set_xlabel('x')

#make the second subplot
axs[0,1].plot(x, np.log(x))

# make the third subplot
axs[1,0].plot(x,y)

#make the fourth subplot
axs[1,1].plot(x, np.tan(x))

plt.tight_layout()
```
